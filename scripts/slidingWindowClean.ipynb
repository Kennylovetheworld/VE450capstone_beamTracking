```python
def instanceIndex(row):
    for i in range(13,26):
        k=j=row[i].find("instance")+len("instance")
        while row[i][k].isnumeric():
            k+=1
        row[i]=row[i][j:k]
```

```python
prev=f[0:1].loc[0].tolist()
# instanceIndex(prev)
prev
```

```python
angle=[]
image=[]
prev=f[0:1].loc[0].tolist()
# instanceIndex(prev)

angle.extend(prev[:13])
image.extend(prev[13:])

for i in range(1,len(f)):
    next=f[i:i+1].loc[i].tolist()
    
    increment=0
    for j in range(25,12,-1):
        if next[j]!=prev[j+increment]:
            increment+=1
            continue
    
    for j in range(12-increment,-1,-1):
        assert(next[j]==prev[j+increment])
    
    angle.extend(next[13-increment:13])
    image.extend(next[26-increment:])
        
    prev=next
```

```python
f.columns
```

```python
print(len(angle))
assert(len(angle)==len(image))
```

```python
cleanData=[]

i=0
while 13*(i+1)<len(angle):
    temp=[]
    temp.extend(angle[13*i:13*i+13])
    temp.extend(image[13*i:13*i+13])
    cleanData.append(temp)
    i+=1

temp=[]
temp.extend(angle[13*i:])
temp.extend(image[13*i:])
cleanData.append(temp)
```

```python
ans=pd.DataFrame(data=cleanData,columns=f.columns)
```

```python
ans.to_csv("slidingWindowClean.csv")
```
